```python
X_train, X_test, y_train, y_test= train_test_split(X , y, test_size=0.25, random_state=42,shuffle=True)
```

```python
Scalar_Standard = StandardScaler()
X_train = Scalar_Standard.fit_transform(X_train)
X_test = Scalar_Standard.transform(X_test)
```

```python
Log_reg = LogisticRegression()
Log_reg.fit(X_train, y_train)
```

```python
confusion_matrix(y_test, Log_reg.predict(X_test))
```

```python
print(classification_report(y_test, Log_reg.predict(X_test)))
```

```python
acuuracy_score = accuracy_score(y_test, Log_reg.predict(X_test))
print(acuuracy_score)
```

```python
error_rate = 1 - acuuracy_score
print(error_rate)
```
