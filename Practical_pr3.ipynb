```python
aggregated_data = df.groupby(['Genre', 'Age']).agg(
    total_salary=('Annual Income (k$)', 'sum'),
    avg_salary=('Annual Income (k$)', 'mean'),
)

print(aggregated_data)
```

```python
aggregateds_data = df.groupby(['Genre', 'Age']).agg(
    total_spend_score=('Spending Score (1-100)', 'sum'),
    avg_spend_score=('Spending Score (1-100)', 'mean'),
)

aggregateds_data
```

```python
bins = [20,30,40,50,60]
labels= ['teen','adult','middle aged','senior']
```

```python
df['Age']=pd.cut(df['Age'],bins=bins,labels=labels,right=False)
```

```python
df.groupby('Age').agg(
    total_salary=('Annual Income (k$)','sum'),
    avg_salary=('Annual Income (k$)','mean'),
    count =('Age','count')
)
```
